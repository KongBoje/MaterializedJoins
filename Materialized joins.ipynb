### 2. We are talent scouts looking to win over some of the best new drivers there are. But we don't want them too old! 
### Write a query that finds the winner of all the races, but only if they are younger than 38 years. 
### The query should return the date, driver surname, driver age, track time in miliseconds, race name and circuit name for all races.

```python
%%sql
SELECT  races.name as race_name, 
        races.date as race_date, 
        date_part('year', age(drivers.dob)) AS driver_age,
        milliseconds, 
        drivers.surname, 
        circuits.name as circuit_name FROM results
        JOIN drivers ON date_part('year', age(drivers.dob)) < 38
            AND results.position = 1
            AND drivers.driverid = results.driverid
        JOIN races USING (raceid)
        JOIN circuits ON circuits.circuitid = races.circuitid;
```

### 4. Create a materialized view of your query. Using EXPLAIN ANALYZE try to query the view.
### Write at least 5 lines of text explaining what's going on and why the query execution time changed.

```python
%%sql
CREATE MATERIALIZED VIEW race_winners_below_38 AS SELECT  races.name as race_name, 
        races.date as race_date, 
        date_part('year', age(drivers.dob)) AS driver_age,
        milliseconds, 
        drivers.surname, 
        circuits.name as circuit_name FROM results
        JOIN drivers ON date_part('year', age(drivers.dob)) < 38
            AND results.position = 1
            AND drivers.driverid = results.driverid
        JOIN races USING (raceid)
        JOIN circuits ON circuits.circuitid = races.circuitid;
```

```python
%sql EXPLAIN ANALYZE SELECT * FROM race_winners_below_38;
```

The query execution time changed because a materialized view creates a static snapshot of the data. this means it only runs one sequential scan through the results, since we already got it.

Planning and execution time in this is then much lower, as seen planning is on 0.100 ms and Exection is on 0.074 ms. If anything changes in the dataset, updated results won't be shown.
